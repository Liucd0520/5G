### **判定模型**

```python
DCE_names = sorted(glob.glob(os.path.join(base_dir, 'DCE', '*.nii.gz')))
label_names = sorted(glob.glob(os.path.join(base_dir, 'DCE_label', '*.nii.gz')))
DCEReg_names = sorted(glob.glob(os.path.join(base_dir, 'DCE_Reg', '*.nii.gz')))
for DCE_name, Reg_name, label_name in zip(DCE_names, DCEReg_names, label_names):
    print(DCE_name)
    if not (DCE_name.split('/')[-1].replace('_0000', '') == \
          Reg_name.split('/')[-1].replace('+C2_Reg_0000', '+C') == \
          label_name.split('/')[-1]):
        print('name: ', DCE_name, Reg_name, label_name)
    
    if not (sitk.ReadImage(DCE_name).GetSize() == \
          sitk.ReadImage(Reg_name).GetSize() == \
          sitk.ReadImage(label_name).GetSize()):
        print('Shape: ', DCE_name, Reg_name, label_name)
    
    if not (sitk.ReadImage(DCE_name).GetSpacing()[1] == \
          sitk.ReadImage(Reg_name).GetSpacing()[1] == \
          sitk.ReadImage(label_name).GetSpacing()[1]):
        print('Spacing 1: ', DCE_name, Reg_name, label_name)
    
    if not round(sitk.ReadImage(DCE_name).GetSpacing()[-1], 1) == \
        round(sitk.ReadImage(Reg_name).GetSpacing()[-1], 1) == \
        round(sitk.ReadImage(label_name).GetSpacing()[-1], 1):
        print('Spacing -1: ', DCE_name, Reg_name, label_name)
        print(sitk.ReadImage(DCE_name).GetSpacing()[-1], 
              sitk.ReadImage(Reg_name).GetSpacing()[-1], 
             sitk.ReadImage(label_name).GetSpacing()[-1])
       
```

```python
import shutil
ids_list = ['1681708', '1354332', '1819369']
for ids in ids_list:
    ADC_path = glob.glob(os.path.join(base_dir, 'ADC', '*' + ids + '*'))[0]
    Reg_path = glob.glob(os.path.join(base_dir, 'ADC_Reg', '*' + ids + '*'))[0]
    label_path = glob.glob(os.path.join(base_dir, 'DWI_label', '*' + ids + '*'))[0]
    print(ADC_path)
    ADC_img = sitk.ReadImage(ADC_path)
    Reg_img = sitk.ReadImage(Reg_path)
    label_img = sitk.ReadImage(label_path)
    
    print(ADC_img.GetSpacing(), Reg_img.GetSpacing(), label_img.GetSpacing())
    ADC_img.SetSpacing(label_img.GetSpacing())
    ADC_img.SetDirection(label_img.GetDirection())
    ADC_img.SetOrigin(label_img.GetOrigin())
    
    Reg_img.SetSpacing(label_img.GetSpacing())
    Reg_img.SetDirection(label_img.GetDirection())
    Reg_img.SetOrigin(label_img.GetOrigin())
    
    sitk.WriteImage(ADC_img, os.path.join('temp', ADC_path.split('/')[-1]))
    sitk.WriteImage(Reg_img, os.path.join('temp', Reg_path.split('/')[-1]))
    
    shutil.copyfile(label_path, os.path.join('temp',label_path.split('/')[-1]))
    
```
